```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ReadCSV") \
    .master("spark://spark-master:7077") \
    .config("spark.executor.cores", "2") \
    .config("spark.cores.max", "4") \
    .config("spark.executor.memory", "2g") \
    .getOrCreate()

spark
```

```python
# Reading CSV - Fligth Data

flight_df = spark.read.format('csv')\
            .option('header', 'false')\
            .option('inferschema', 'false')\
            .option('mode','FAILFAST')\
            .load('/opt/spark-data/input/flight_data.csv')

flight_df.show(5)
```

```python
# Reading CSV - Fligth Data w/ header

flight_header_df = spark.read.format('csv')\
            .option('header', 'true')\
            .option('inferschema', 'false')\
            .option('mode','FAILFAST')\
            .load('/opt/spark-data/input/flight_data.csv')

flight_header_df.show(5)
```

```python
flight_header_df.printSchema()
```

```python
# Reading CSV - Fligth Data w/ header w/ schema

flight_schema_df = spark.read.format('csv')\
            .option('header', 'true')\
            .option('inferschema', 'true')\
            .option('mode','FAILFAST')\
            .load('/opt/spark-data/input/flight_data.csv')

flight_schema_df.printSchema()
```
